```python
import numpy as np
import cv2
from glob import glob
import matplotlib.pyplot as plt
```

```python
female_file_path = glob('data/female/*.jpg')
male_file_path = glob('data/male/*.jpg')
```

```python
# Print the number of images in each dataset
print('The number of images in female folder:', len(female_file_path))
print('The number of images in male folder:', len(male_file_path))
```

## Crop all `female` images

```python
face_cascade = cv2.CascadeClassifier('model/haarcascade_frontalface_default.xml')
```

```python
def crop_faces(image_path, genderName):
    for i in range(len(image_path)):
        try: 
            # Read images and convert to RGB
            img = cv2.imread(image_path[i])
            img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

            # Apply haar cascade to detect faces
            gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
            faces = face_cascade.detectMultiScale(gray, 1.5, 5)

            # Print the bounding box of each detected face
            for (x,y,w,h) in faces:
                # Crop the faces and resize to 64x64
                crop_img = img[y:y+h, x:x+w]
                # Save the cropped faces
                cv2.imwrite(f'./crop_data/{genderName}/{genderName}_{i}.jpg', crop_img)
                print(f'Image {i} is processed successfully')
        except:
            print('Unable to process the image')
```

```python
crop_faces(female_file_path, 'female')
```

## Crop all `male` images

```python
crop_faces(male_file_path, 'male')
```
