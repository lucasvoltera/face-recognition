```python
from pipeline import process_image, display_original_image, display_prediction_results
import cv2
import matplotlib.pyplot as plt
```

# Face Recognition Prediction

```python
file_path = 'test_images/getty_test.jpg'

pred_img, pred_dict = process_image(file_path)
display_original_image(pred_img)
display_prediction_results(pred_dict)
```

```python
def run_video_prediction(video_path):
    cap = cv2.VideoCapture(video_path)

    while True:
        ret, frame = cap.read()

        if not ret:
            break

        pred_img, pred_dict = process_image(frame, path=False)

        cv2.imshow('Prediction', pred_img)

        if cv2.waitKey(1) == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()

video_path = 'test_images/video.mp4'
run_video_prediction(video_path)
```
